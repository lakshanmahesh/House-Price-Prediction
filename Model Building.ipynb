```python
data = data.drop(columns=['basement','prefarea'])
```

```python
data = pd.get_dummies(data)
```

```python
X = data.drop('price',axis = 1)
y= data['price']
```

```python
pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
```

```python
X_train.shape,X_test.shape
```

```python
def model_acc(model):
    model.fit(X_train,y_train)
    acc = model.score(X_test,y_test)
    print(str(model)+ ' --> ' +str(acc))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model_acc(lr)

from sklearn.linear_model import Lasso
lasso = Lasso()
model_acc(lasso)

from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
model_acc(dt)

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
model_acc(rf)
```

```python
from sklearn.model_selection import RandomizedSearchCV

param_distributions = {
    'fit_intercept': [True, False],
    'positive': [True, False],
    'copy_X': [True, False]
}

random_search = RandomizedSearchCV(estimator=lr,param_distributions=param_distributions,
                                   n_iter=5,cv=5,random_state=42
)
random_fit = random_search.fit(X_train, y_train)
best_model = random_fit.best_estimator_
best_model
```

```python
best_model.score(X_test,y_test)
```

```python
import pickle
with open('predictor.pickle','wb') as file:
    pickle.dump(best_model,file)
```
